# Importing Libraries for Data Analysis

```python
##Importing Libraries for Data Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Loading the Dataset

```python
# I have downloaded the Dataset from Kaggle website for this project.      Link - https://www.kaggle.com/datasets/mojtaba142/hotel-booking
# Our data is in the csv format
# Importing data for our analysis

df = pd.read_csv('hotel_booking.csv')
```

# Data Cleaning

```python
#Checking how many rows and columns are present in our data
df.shape
```

```python
df.head()  #gives first 5 rows of data
```

## Removing unnecessary columns

```python
# Analysed Data and Removing unwanted columns like 'name', 'email', 'phone-number', 'credit_card number' from dataset for analysis

df.drop('name', axis=1, inplace=True)
df.drop('email', axis=1, inplace=True)
df.drop('phone-number', axis=1, inplace=True)
df.drop('credit_card', axis=1, inplace=True)
```

```python
df.shape
```

```python
df.columns
```

## Changing DataType

```python
df.info()
```

```python
# Changing the datatype of 'reservation_status_date' to datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

## Removing NULL values from Data

```python
#Checking for null values in data
df.isnull().sum()
```

```python
# As 'agent' and 'company' columns has to many null values so droping these columns for better analysis
df.drop(['agent','company'], axis=1, inplace=True)

# Removing all null values
df.dropna(inplace=True)


df.isnull().sum()
```

## Removing Outliers From Data

```python
df.describe()
```

```python
# From above it is evident that 'adr' column i.e. 'Average Daily Rate' has outliers.
# It's min = -6.380000 and max = 5400.0000000
# We can visualise this in the below box plot

df['adr'].plot(kind = 'box')
```

```python
# As we can see from the above box plot values for 'adr' column range below 1000
# So filtering data of 'adr' column values below 1000

df=df[df['adr']<1000]
df.describe()
```

# Data Analysis and Visualizations

```python
# Research Questions

#  1.  What are the variables that affect hotel reservation cancellations?
# 2.  How can we make hotel reservations cancellations better?
# 3.  How will hotels be assisted in making pricing and promotional decisions?
```

```python
# How many bookings are canceled in terms of percentage
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

#Ploting Bar Graph
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor='k', width = 0.7)
plt.show()
```

```python
# Depending on the hotel which hotel has the highest cancelation

plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Checking if Price has any effect on Booking Cancelations in these Hotels

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
# Filtering Data for dates between 2016 and 2018

resort_hotel = resort_hotel[(resort_hotel.index >= '2016-01-01') & (resort_hotel.index <= '2018-12-31')]
city_hotel = city_hotel[(city_hotel.index >= '2016-01-01') & (city_hotel.index <= '2018-12-31')]

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] =  df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title( 'Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('Number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index(), palette = 'husl')
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled']== 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')

plt.xlabel("Reservation Status Date")
plt.ylabel("ADR")
plt.legend()

plt.show()
```

```python
# Filtering Data For 'reservation_status_date' between '2016' and '2017-09' as per hypothesis and to provide more information.

cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016') &  (cancelled_df_adr['reservation_status_date'] < '2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016') &  (not_cancelled_df_adr['reservation_status_date'] < '2017-09')]


# Plotting the graph

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')

plt.xlabel("Reservation Status Date")
plt.ylabel("ADR")
plt.legend()

plt.show()
```
